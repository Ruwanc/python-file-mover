```python
import os

import datetime, shutil
from shutil import move, Error

#video_ext = ['.mp4', '.m4a', '.m4v', '.f4v', '.f4a', '.m4b', '.m4r', '.f4b', '.mov', '.wmv', '.wma', '.avi']
#document_ext = ['.txt', '.pdf', '.docx']

# print (now.strftime("%Y-%m-%d"))
# check if path exists
def check_path(path):
    if os.path.exists(path):
        return True
    else:
        return False

#check path exists if not create it
def path_check_create(path):
    
    if not os.path.exists(path):
        
        try:
            os.makedirs(path)
            print ('Path created')
            return True
        except OSError as error:
            print(error)
            return False
        
def move_file(current_file_src, original_dst, key):
    final_dst = original_dst+"\\"+key
    try:
        move(current_file_src, final_dst)
    except Error as e:
        print(e)
        
        
### file_move function begin
def file_mover(src_path, dst_path):
    file_count = 0
    key_video = "Video"
    key_document = "Document"
    video_ext = ['.mp4', '.m4a', '.m4v', '.f4v', '.f4a', '.m4b', '.m4r', '.f4b', '.mov', '.wmv', '.wma', '.avi']
    document_ext = ['.txt', '.pdf', '.docx']
    #check wheather the soure path is exists
    if check_path(src_path):
        # check file directory is not-empty
        if os.listdir(src_path+"\\Video"):
            for r, d, f in os.walk(src_path+"\\Video"):
                for file in f:
                    # check the files in video folder
                    temp_src = src_path+"\\Video\\"+file
                    filename,file_extension = os.path.splitext(temp_src)
                    
                    if file_extension in video_ext:
                        temp_dst = dst_path+"\\Video"
                        #check the availability of temporary destination path the if not create it
                        path_check_create(temp_dst)
                        move_file(temp_src,dst_path,key_video)
                        file_count = file_count + 1
                        print(file+" moving process complete")
                        continue
                        
                    if file_extension in document_ext:
                        temp_dst = dst_path+"\\Document"
                        #check the availability of temporary destination path the if not create it
                        path_check_create(temp_dst)
                        move_file(temp_src,dst_path,key_document)
                        file_count = file_count + 1
                        print(file+" moving process complete")
                        
        else:
            print(src_path+" is EMPTY")
                        
    else:
        print(src_path+" path does not exists") #source path not exists
            
    
    
def main():
    src_path = 'C:\\Users\\Ruwan\\Desktop\\test'
    dst_path = 'E:\\Downloads'

    file_mover(src_path, dst_path)
    
if __name__ == '__main__':
    main() #call the main function
```
